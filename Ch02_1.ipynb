```python
img = cv2.imread('/content/gdrive/My Drive/Ch01/IP/Images/lena2.jpg')
print(img.shape)
```

```python
cv2_imshow(img)
```

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #색상 설정
```

```python
plt.imshow(img) #그냥 img로만 입력하면 비트맵에 저장되어있는 그대로 뜸
```

```python
plt.imshow(RGB_img) #RGB_img를 해야 정상적인 컬러 이미지로 나온다.
```

```python
R_img, G_img, B_img = cv2.split(RGB_img) #채널분리
```

```python
print(R_img.shape)
```

```python
CImg = RGB_img
```

```python
CImg[:,:,0] = R_img  #R(빨간색)에 대한 정보를 띄어주는 코드
CImg[:,:,1] = R_img
CImg[:,:,2] = R_img
```

```python
plt.imshow(CImg) #빨간색의 대해 보여주는것
```

```python
CImg[:,:,0] = G_img  # G(초록색)에 대한 정보를 띄어주는 코드
CImg[:,:,1] = G_img
CImg[:,:,2] = G_img
```

```python
plt.imshow(CImg)
```

```python
CImg[:,:,0] = B_img #B(파란색)에 대한 정보를 띄어주는 코드
CImg[:,:,1] = B_img
CImg[:,:,2] = B_img
```

```python
plt.imshow(CImg)
```

```python
HSV_img = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2HSV)
```

```python
H_img, S_img, V_img = cv2.split(HSV_img) #채널분리
CImg[:,:,0] = H_img
CImg[:,:,1] = H_img
CImg[:,:,2] = H_img
plt.imshow(CImg)
```

```python
H_img, S_img, V_img = cv2.split(HSV_img) #채널분리
CImg[:,:,0] = S_img
CImg[:,:,1] = S_img
CImg[:,:,2] = S_img
plt.imshow(CImg)
```

```python
H_img, S_img, V_img = cv2.split(HSV_img) #채널분리
CImg[:,:,0] = V_img
CImg[:,:,1] = V_img
CImg[:,:,2] = V_img
plt.imshow(CImg)
```
