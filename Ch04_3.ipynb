```python
from google.colab.patches import cv2_imshow
from google.colab import drive
import numpy as np # numpy는 배열 사용할 때
import cv2
import matplotlib.pyplot as plt
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/Ch01/IP/Images/girl.jpg')
print(img.shape)
gImg = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
# plt.imshow(gImg, cmap='gray')
cv2_imshow(gImg)
```

```python
hist = cv2.calcHist([gImg], [0], None, [256], [0, 256])
plt.plot(hist)
```

```python
outImg = np.zeros(gImg.shape, dtype='uint8')
```

```python
min = 0
max = 200

for h in range(gImg.shape[0]):
  for w in range(gImg.shape[1]):
    if gImg[h ,w] < min:
      outImg[h, w] = 0
    elif gImg[h, w] > max:
      outImg[h, w] = 255
    else:
        val = np.int32(gImg[h, w] - min) * 255
        val = np.int32(val / (max - min))
        outImg[h, w] = val

cv2_imshow(outImg)
```

```python
hist2 = cv2.calcHist([outImg], [0], None, [256], [0, 256])
plt.plot(hist2)
```
