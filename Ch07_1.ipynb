```python
from google.colab.patches import cv2_imshow
from google.colab import drive
import numpy as np # numpy는 배열 사용할 때
import cv2
import matplotlib.pyplot as plt
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/ImageProcessing/Ch01/IP/Images/lena256.jpg')
print(img.shape)
gImg1 = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gImg1, cmap='gray')
```

```python
outImg = np.zeros(gImg1.shape, dtype = np.uint8)
cv2_imshow(outImg)
```

```python
scaling = 3

for h in range(gImg1.shape[0]):
  for w in range(gImg1.shape[1]):
    y = int(h * scaling)
    x = int(w * scaling)
    if y < gImg1.shape[0] and  x < gImg1.shape[1]:
     outImg[y, x] = gImg1[h, w]

cv2_imshow(outImg)
```

```python
scaling = 4

for y in range(gImg1.shape[0]):
  for x in range(gImg1.shape[1]):
    h = int(y / scaling)
    w = int(x / scaling)
    if h < gImg1.shape[0] and  w < gImg1.shape[1]:
      outImg[y, x] = gImg1[h, w]

  cv2_imshow(outImg)
```

```python
img = cv2.imread('/content/gdrive/My Drive/ImageProcessing/Ch01/IP/Images/lena256.jpg')
print(img.shape)
gImg3 = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gImg3, cmap='gray')
```

```python
height, width = gImg3.shape
print(height, width)
```

```python
M = cv2.getRotationMatrix2D((width / 2.0, height / 2.0), 45, 1)
print(M)
```

```python
outImage = cv2.warpAffine(gImg3, M, (width, height))
cv2_imshow(outImage)
```
